##인덱스가 순서대로 설정
##1. pd.concat([df1, df2]), ignore_index=True


##옆으로 합치는 방법
##2. pd.concat([df1, df2]), axis=1)

```python
import pandas as pd

english_df = pd.DataFrame({
    'name' : ['dongwook', 'sineui', 'ikjoong', 'yoonsoo'],
    'english_score' : [50, 89, 68, 88]
})

english_df2 = pd.DataFrame({
    'name' : ['yoonsoo', 'sowon', 'seungheon', 'liam'],
    'english_score' : [50, 89, 68, 88]
})
```

```python
pd.concat([english_df, english_df2])
```

```python
#인덱스가 순서대로 설정해보자
english_df = pd.concat([english_df, english_df2], ignore_index=True)
```

```python
#학생들의 수학점수의 데이터프레임
math_df = pd.DataFrame({
    'name' : ['dongwook', 'sineui', 'ikjoong', 'yoonsoo','yoonsoo', 'sowon','seungheon','liam'],
    'math_score' : [50, 89, 68, 88, 100, 100, 79,90]
})
```

```python
math_df
```

```python
#학생들의 영어점수와 수학점수를 합쳐보자,concat함수는 기본적으로 아래에 데이터를 합침
pd.concat([english_df, math_df])
```

```python
#가로로 붙여보자
pd.concat([english_df, math_df],axis=1)
```

#join()연산

##1.inner join : 두 데이터의 공통된데이터들만 합쳐줌

##2.left outer join : 왼쪽데이터를 기준으로 데이터를 합치고, 공통이 아닌 데이터들은 결측값으로 표시

##3.Right outer join : 오른쪽데이터를 기준으로 합치고, 고통이 아닌 데이터들은 결측값으로 표시

##4.Full outer join : 모든 데이터들을 합치고 제외된 데이터들은 결측값의 표시

```python
df = pd.read_csv('/content/drive/MyDrive/코드잇csv/employee.csv')
```

```python
df2 = pd.read_csv('/content/drive/MyDrive/코드잇csv/survey.csv')
```

```python
df
```

```python
df2
```

```python
#merge함수 활용하자, 기본적으로 인어조인을 활용!
#어떤 값들을 키값으로  설정할까? 중복될 가능성이 없는 사원번호를 키값으로 하는것이 좋다
#설문조사를 한 대상들만 불러옴
pd.merge(df, df2, on='id')
```

```python
#설문조사를 시행하지않는 대상들도 포함(전체직원 정보)
pd.merge(df, df2, on='id', how='left')
```

```python
pd.merge(df, df2, on='id', how='right')
```

```python
#같은컬럼들의 이름을 바꾸자, 뒤에 _x, y 값들을 변경하자
pd.merge(df, df2, on='id', how='left', suffixes=['', '_x'])
```

```python
#다른상황에서는?
df2 = df2.rename(columns={'id':'employee_id'})
df2
```

```python
#키값의 컬럼들의 이름이 다르면?
pd.merge(df, df2, left_on='id', right_on='employee_id')
```

```python
company_df = pd.merge(df, df2, left_on='id', right_on='employee_id')
```

```python
company_df
```

#인덱스 기준으로 데이터 합치기: join()

```python
df = pd.read_csv('/content/drive/MyDrive/코드잇csv/employee.csv')
```

```python
df2 = pd.read_csv('/content/drive/MyDrive/코드잇csv/survey.csv')
```

```python
df
```

```python
df2
```

```python
#각각의 사원아이디를 인덱스로 설정하자
df = df.set_index('id')
df2 = df2.set_index('id')
```

```python
#merge()는 컬럼에 겹치는 이름들이 있으면 자동으로 뒤에 _x,_y를 해주지만
#join()는 따로 suffix 값을 설정해야 한다
```

```python
#df는 왼쪽데이터에, df2는 오른쪽에 합쳐지게 됨
df.join(df2, lsuffix='_x', rsuffix='_y')
```

```python
#인어조인을 통해 설문조사를 진행한 데이터들만 불러오자
df.join(df2, lsuffix='_x', rsuffix='_y',how='inner')
```

```python
#위 데이터를 df에 저장하면 2개의 데이터를 함친 데이터가 나타남
df= df.join(df2, lsuffix='_x', rsuffix='_y',how='inner')
```

```python
df
```
