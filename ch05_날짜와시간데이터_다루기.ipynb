```python
#이번에는 주문시간과 배송시간을 통해 실제로 몇일 걸렸는지 계산해보자
df
```

```python
#오른쪽값은 배송소요 시간
df['shipping_time'] - df.index
```

```python
#위 값을 새로운컬럼에 저장하자
df['days_for_shipping(actual)'] = df['shipping_time'] - df.index
```

```python
#잘 저장됬는지 확인해보자
df
```

```python
df.dtypes #예상시간과 실제시간의 타입이 다르면 오류가 날 수 있으니 같은타입으로 변경하자
```

```python
#일을 기준으로 타임델타를 변경하자
pd.to_timedelta(df['days_for_shipping(expected)'],unit='D')
```

```python
#해당값을 저장하고 확인해보자
df['days_for_shipping(expected)'] = pd.to_timedelta(df['days_for_shipping(expected)'],unit='D')
```

```python
df.dtypes #인트였던 타입이 타임델타로 잘 변경된것을 확인
```

```python
#이제 예상배송시간보다 오래걸린 데이터들을 불러오자
df['days_for_shipping(expected)'] < df['days_for_shipping(actual)']
```

```python
#해당값들을 새로운 컬럼에 저장하자
df['late_shipping'] = df['days_for_shipping(expected)'] < df['days_for_shipping(actual)']
df
```

```python
#예상보다 늦게 배송된 건들이 몇개인지 확인해보자 : 3204개
df['late_shipping'].value_counts()
```

##이번에는 배달 음식이 고객에게 도착한 시각을 계산해 봅시다. 주문 시각(order_time)과 배달 소요 시간(time_taken)을 가지고 배달 완료 시간을 계산해서 DataFrame에 delivery_time이라는 컬럼을 추가해 주세요!
#참고로 time_taken 컬럼은 배달에 몇 분이 소요되었는지를 의미하므로, 아래 출력 결과처럼 분을 기준으로 한 Timedelta로 바꿔 주세요. 참고로 분을 의미하는 코드는 T 입니다.

```python
delivery_df
```

```python
#타입을 먼저 알아보자
delivery_df.dtypes
```

```python
#taken타입을 타임델타로 변경하자
delivery_df['time_taken'] = pd.to_timedelta(delivery_df['time_taken'],unit='T')
```

```python
#도착한 시간을 계산하는 방법은? 그리고 새로운 컬럼에 저장하자(주문시간과 소요시간을 더하자)
delivery_df['delivery_time'] = delivery_df.index + delivery_df['time_taken']
```
