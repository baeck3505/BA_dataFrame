###이전 영상에서 여러 개의 조건을 가지고 불린 인덱싱을 하는 방법에 대해 알아봤는데요. 이번 레슨에서는 불린 인덱싱을 할 때 유용하게 사용할 수 있는 query() 함수를 소개해 드리려고 합니다.

###일단 예를 들어 '대출 신청자의 수입이 5000 이상이다'를 의미하는 조건식 1개를 가지고 불린 인덱싱을 해 볼게요. 지금까지 배운 방법을 사용하면 아래와 같이 DataFrame 뒤에 대괄호를 쓰고 그 안에 조건식을 넣어 주면 됩니다.

###그런데 DataFrame에 query() 함수를 사용하면 좀 더 간단한 코드로 동일한 결과물을 얻을 수 있어요. 아래와 같이 query() 함수에 따옴표(' 혹은 ") 안에 조건식을 넣어서 값을 넘겨주면 되는데요. 조건식에서 DataFrame에 대한 언급을 생략하고 바로 컬럼 이름만 쓰면 되어서 코드가 좀 더 간결해지는 효과가 있습니다.

loan_df.query('income > 5000')


###이때 조건식에 변수를 사용하는 것도 가능합니다. 예를 들어 income_mean이라는 변수에 대출 신청자들의 수입 평균을 저장해 두었다고 할게요. 그리고 query() 안에 'income > @income_mean' 이런 식으로 써 주면 수입이 평균보다 큰 신청자들의 데이터만 뽑을 수 있습니다. 이때 변수 앞에 @를 쓰지 않으면 오류가 나니까 주의해 주세요!

income_mean = loan_df['income'].mean()

loan_df.query('income > @income_mean')


###그렇다면 조건식 안에 문자열을 넣어야 할 때에는 어떻게 하면 될까요? 예를 들어 지난 레슨에서 만든 조건식처럼 '대출 신청자의 담보 부동산이 도시에 있다'를 의미하는 조건식을 사용하려면, 아래 코드처럼 큰 따옴표(")와 작은 따옴표(')를 함께 활용해야 합니다. 참고로 한 가지 더 알려 드리자면, 컬럼명에 공백이 포함되어 있으면 property area 이런 식으로 컬럼명을 백틱(```)으로 감싸 주면 됩니다. 어렵지 않죠?

loan_df.query("property_area == 'City'")


###그리고 이때 등호나 부등호로 된 식 말고도, True나 False 값을 리턴하는 다양한 형태의 조건식을 사용할 수 있어요. 그 중에서도 in이나 not in 연산자를 많이 활용하는 편인데요. 예를 들어 아래와 같은 코드를 통해 담보 부동산이 City나 Suburb에 해당하는 데이터를 추출할 수 있어요.

loan_df.query("property_area in ['City', 'Suburb']")
loan_df.query("property_area not in ['City', 'Suburb']")


###그리고 query() 함수에서도 마찬가지로 여러 개의 조건을 결합해서 사용할 수 있는데요. 예를 들어 아래와 같이 and(&) 연산자를 사용해서 기혼 상태이고, 수입이 평균보다 많은 신청자들의 데이터를 뽑을 수 있습니다. 물론 not(~)이나 or(|) 연산자도 활용할 수 있고요.

loan_df.query("married == 'Y' & income > @income_mean")


###각각의 조건식을 좀 더 명확하게 볼 수 있도록 괄호로 묶어 줄 수도 있습니다. 특히 not(~) 연산자를 사용하는 경우에는 괄호로 조건식을 묶어 줘야 코드가 예상치 못하게 작동하거나, 오류가 발생하는 걸 방지할 수 있어요.

loan_df.query("(married == 'Y') & (income > @income_mean)")

##지난 레슨에서는 drop() 함수를 사용해서 컬럼과 로우를 삭제하는 방법에 대해 알아봤습니다. 로우를 삭제할 때는 함수 안에 바로 삭제할 로우의 인덱스 값을 넣어 주면 되고, 컬럼을 삭제할 때는 columns라는 파라미터를 사용해서 삭제할 컬럼의 이름을 넣어 주면 된다고 배웠었죠?

##그런데 drop() 함수를 쓸 때에는 axis라는 파라미터도 많이 사용합니다. axis는 연산을 컬럼 방향으로 할지, 로우 방향으로 할지 정할 수 있는 함수에서 주로 사용되는데요. drop()에서는 이 값에 따라 로우를 삭제할지, 컬럼을 삭제할지 정할 수 있어요.

##일단 drop()에서 axis의 기본값은 로우를 삭제하는 것을 의미하는 0입니다. 즉, 아래와 같이 loan_df에서 인덱스가 LP1006인 로우를 삭제할 때 함수 내부적으로는 loan_df.drop('LP1006', axis=0)처럼 작동하는 거죠.

##참고로 axis='index'라고 써 줘도 똑같이 인덱스 이름을 기준으로 로우를 삭제하게 됩니다. 하지만 어차피 기본값이 로우를 삭제하는 거니까 이런 경우엔 굳이 axis 값을 따로 설정할 필요가 없겠죠?

##그다음으로, 컬럼을 삭제할 땐 axis 값을 1로 설정하면 됩니다. 예를 들어 이전 영상처럼 loan_df에서 married 컬럼을 삭제하려면 아래와 같이 코드를 작성하면 되죠. 그리고 axis=0과 axis='index'가 똑같이 작동하는 것처럼 axis=1과 axis='columns'도 똑같다고 보시면 됩니다.

##즉 컬럼을 삭제할 때는 drop()에서 columns 값에 바로 컬럼명을 넣을 수도 있고, axis 값을 1이나 columns로 줄 수도 있습니다. 두 방법 모두 같은 결과물을 얻을 수 있지만 전자가 코드도 좀 더 짧고, 이해하기 쉽다는 장점이 있어서 영상에서는 columns 파라미터를 사용하는 방식을 설명드렸는데요. 물론 axis를 사용하는 게 더 편하신 분들은 이 방법을 사용하시면 됩니다.
