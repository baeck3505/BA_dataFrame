##pandas에는 category라는 데이터 타입이 있습니다. category가 영어로 ‘범주’를 의미하는 것에서 알 수 있듯이, category 타입은 범주형 데이터를 표현할 때 사용하는데요. 범주형 데이터는 연속적인 값을 갖는 연속형 데이터와 달리, 보통 어느 정도 제한된 범위 안에서 몇 가지 값을 가집니다. 예를 들어, 옷 사이즈는 XS/S/M/L/XL, 혈액형은 A형/B형/O형/AB형, 연령대는 10대/20대/30대 등의 범주를 가지는 것처럼요.

##지난 영상에서 다룬 데이터를 예로 들면, genre 컬럼은 Comedy, Documentary, Drama, Romantic comedy, Thriller 이렇게 다섯 종류의 값을 가진 범주형 데이터로 되어 있다고 할 수 있어요. 그런데 엄연히 말하자면, genre 컬럼의 데이터 타입은 object입니다. 텍스트로 되어 있는 데이터는 pandas가 기본적으로 object 타입으로 불러오죠.

##그렇다면 object 타입이 있는데 굳이 category 타입이 존재하는 이유는 뭘까요? 먼저 category 타입은 좀 더 메모리 사용량이 적다는 특징이 있고요. 그리고 가장 중요한 특징으로는 각 범주에 순서를 매길 수 있다는 점을 들 수 있습니다. 일단 예를 들어 아래와 같이 옷 사이즈와 판매량 정보가 들어 있는 DataFrame이 있다고 합시다.

```python
import pandas as pd

clothes_df = pd.DataFrame({
        'size': ['L', 'S', 'XS', 'L', 'S', 'XL', 'L', 'S', 'M', 'XS',
             'M', 'M', 'XS', 'L', 'XL', 'XS', 'M', 'S', 'L', 'XL'],
    'sales': [130, 200, 120, 120, 140, 160, 190, 90, 110, 100,
              150, 180, 100, 200, 80, 140, 150, 90, 80, 130]
})

clothes_df

##오름 차순을 하면 안된다! 한글이 아니라 알파벳이기 때문에! 원하는대로 순서를 정할려면?
##카테고리 함수를 사용하면 좋다


pd.Categorical(clothes_df['size'], ordered=True, categories=['XS', 'S', 'M', 'L', 'XL'])

clothes_df['size'] = pd.Categorical(clothes_df['size'],
                                              ordered=True,
                                                                    categories=['XS', 'S', 'M', 'L', 'XL'])


clothes_df.sort_values(by='size') #잘 정리되서 나옴
```

#groupby와 멀티 인덱스

```python
df = pd.read_csv('/content/drive/MyDrive/코드잇csv/netflix.csv')
```

```python
df
```

```python
#개봉일의 날짜값만 불러오자
pd.to_datetime(df['release'])
```

```python
#년도 값만 불러오자
pd.to_datetime(df['release']).dt.year
```

```python
#새로운 컬럼에 저장하자
df['year'] = pd.to_datetime(df['release']).dt.year
```

```python
df
```

```python
df.dtypes
```

```python
df.groupby(['genre','year'])
```

```python
#장르와 년도를 가지고 그룹바이 해보고 평균값과 숫자값들만 불러오자
##이것을 멀티 인덱스
df.groupby(['genre','year']).mean(numeric_only=True)
```

```python
#반대로도 가능함
#year처럼 바깥쪽에 있는 인덱스는 하나씩 접근이 가능하다
##안쪽에 있는 장르에 바로 접근하는 것은 어렵다
df.groupby(['year','genre']).mean(numeric_only=True)
```

```python
#2020년의 정보만 가져오자
df.groupby(['year','genre']).mean(numeric_only=True).loc[2020, :]
```

```python
#2020년 드라마의 평균평점과 평균사영시간은?
##loc 함수에 리스트도 가능
df.groupby(['year','genre']).mean(numeric_only=True).loc[2020, 'Drama']
```

```python
#튜플안에서 리스트를 넘겨줄때는 콤마와 콜론을 써서 앞부분은 인덱스를, 귓부분에는 컬럼이라는 것을 지정해줘야 함)
df.groupby(['year','genre']).mean(numeric_only=True).loc[(2020, ['Drama','Comedy']), :]
```

```python
#모든년도의 드라마 장르의 정보를 가져올려면?
df.groupby(['year','genre']).mean(numeric_only=True).loc[([2019, 2020, 2021], 'Drama'), :]
```

```python
#다른방법으로는?
years = sorted(df['year'].unique())
```

```python
df.groupby(['year','genre']).mean(numeric_only=True).loc[(years, 'Drama'), :]
```

#실습 설명
##피벗 테이블로 몬트리올 올림픽의 국가별 금메달, 은메달, 동메달의 개수를 계산해 봅시다. 이때 결측값은 0으로 채우고, 금메달, 은메달, 동메달 개수, 나라 이름 순서대로 내림차순 정렬해 주세요. 메달 정보는 medal 컬럼에 저장되어 있고, 1st는 금메달, 2nd는 은메달, 3rd는 동메달을 의미합니다. 정렬 방식에 대한 자세한 정보는 아래를 확인해 주세요!

##금메달 개수를 기준으로 먼저 내림차순 정렬합니다.
##금메달 개수가 같은 경우에는 은메달 개수를 기준으로 내림차순 정렬합니다.
##금메달과 은메달 개수도 같은 경우에는 동메달 개수를 기준으로 내림차순 정렬합니다.
##금메달, 은메달, 동메달 개수도 모두 같은 경우에는 나라 이름을 기준으로 내림차순 정렬합니다.
##이렇게 순서대로 값을 정렬하고 싶을 때에는 by 파라미터에 리스트를 넘겨주면 되는데요. 리스트 안에 기준이 되는 컬럼들을 순서대로 써 주면 됩니다.

```python
import pandas as pd

olympic_df = pd.read_csv('data/olympic.csv')

pd.pivot_table(
    olympic_df,
    index='team',
    columns='medal',
    aggfunc='size'
).fillna(0).sort_values(
    by=['1st', '2nd', '3rd', 'team'],
    ascending=False
)
```
