###1. brand 컬럼에 저장된 제조사명의 대소문자 표기를 통일해 주세요. 첫 글자는 대문자로, 나머지 글자는 소문자로 바꿔 주시면 됩니다.


###2.name 컬럼에는 iPhone 14 Pro (256GB)와 같이 스마트폰의 모델명(iPhone 14 Pro)과 용량(256GB) 정보가 함께 들어 있습니다. 모델과 용량을 쉽게 구분해서 볼 수 있도록 문자열을 분리해 주세요. 모델명은 model 컬럼, 용량은 capacity 컬럼에 저장하고, 기존의 name 컬럼은 삭제하면 됩니다. 모델명 앞뒤에 불필요한 공백 문자열이 없도록 처리하는 것도 잊지 마세요!


###3. size 컬럼에는 스마트폰의 디스플레이 크기 정보가 담겨 있습니다. 그런데 데이터에 인치(inch)를 나타내는 " 기호가 들어 있어서 pandas가 이 컬럼의 데이터 타입을 문자 데이터로 인식하고 있네요. " 기호를 없앤 뒤 size 컬럼의 데이터 타입을 적절한 숫자 타입으로 바꿔 주세요.

```python
cellphone_df = pd.read_csv('/content/drive/MyDrive/코드잇csv/cellphone.csv')
cellphone_df
```

```python
# 첫글자 대문자로 통일
cellphone_df['brand'] = cellphone_df['brand'].str.capitalize()
```

```python
cellphone_df
```

```python
cellphone_df['model'] = cellphone_df['name'].str.split('(').str[0] #(를 기준으로 나눠서 새로운 모델컬럼에 저장
```

```python
cellphone_df['capecity'] = cellphone_df['name'].str.split('(').str[1] #(를 기준으로 나눠서 새로운 용량컬럼에 저장
```

```python
cellphone_df['model'] = cellphone_df['model'].str.strip() #모델컬럼에 데이터의 앞뒤 공백을 제거
```

```python
cellphone_df['capecity'] = cellphone_df['capecity'].str.replace(')','') #용량데이이터의 끝 부분 )를 공백으로 바꾸고 저장
```

```python
cellphone_df.drop(columns = 'name') #네임컬럼을 삭제
```

```python
cellphone_df
```

```python
cellphone_df['size'] = cellphone_df['size'].str.replace('"','').astype('float') #사이즈시리즈 뒤에 "를 제거하고 소수형으로 변환 후 저장
```

```python
cellphone_df
```
