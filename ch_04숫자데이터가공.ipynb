#실습 설명
##비만도를 나타내는 지표로 'BMI(Body Mass Index)'라는 것도 많이 사용합니다. BMI를 계산하려면 몸무게(kg)를 키(m)의 제곱으로 나눠 주면 되는데요. 예를 들어, 키가 180cm이고, 몸무게가 70kg인 사람의 BMI는 70을 1.8의 제곱으로 나눈 값인 21.6입니다.

##위 내용을 참고해서 patient_df에 있는 각 환자의 BMI 값을 계산해서 bmi라는 컬럼에 저장해 주세요. 단, BMI 값은 소수 첫째 자리까지 반올림해 주셔야 합니다.



```python
df
```

```python
df['height'] **2 #키의 제곱
```

```python
df['weight'] / df['height'] **2 #비만도를 나타내는 BMI : 몸무게를 키의 제곱으로 나눈 값
```

##스케일링(scaling) :  숫자 데이터의 단위나 범위를 일정하게 맞추는 것
##왜 필요할까? 값을 표혀하는 단위가 다르면 수치화 하기가 어렵다. 분석결과 안좋은 영향을 줄 수 있음
##정규화 : 데이터를 0과 1사이 값으로 바꾸는 것(최소값과 최댓값을 사용), 데이터의 범위를 통일하고 싶을 때 사용

```python
df
```

```python
#환자들의 키값들을 정규화해보자
(df['height'] - df['height'].min()) / (df['height'].max() - df['height'].min())
```

```python
df['height'] = (df['height'] - df['height'].min()) / (df['height'].max() - df['height'].min())
```

```python
df
```

```python
df['height'].describe() #정규화가 제대로 되었는지 확인해보자
```

#표준화
##각 값이 평균에 비해 얼마나 크거나 작은지 나타내는 것으로 평균은 0으로, 분산은 1로 변환
## 평균과 표준편차 값을 알야아 함
## 이렇게 구한 값을 z- score라고 함. 데이터가 평균으로부터 얼마나 떨어져 있는지를 나타내는 값으로 이 때 평균과의 거리를 표준편차 기준으로 표현함
##예를 들어 z-score가 1.5이면 평균보다 1.5표준편차보다 값이 큰 값
##즉 데이터들이 평균에서 얼마나 떨어져있는지를 비교하고 싶을 때 사용

```python
df
```

```python
#키의 값을 표준화 해보자
(df['height'] - df['height'].mean()) / df['height'].std()
```

```python
#키의 컬럼에 저장하자
df['height'] = (df['height'] - df['height'].mean()) / df['height'].std()
```

```python
df
```

```python
#표준화가 잘 되어있는지 확인하자
df['height'].describe()
```

#cut() 함수로 데이터 구간화하기(binning)
##연속적인 숫자 데이터를 여러개의 구간으로 분류하는 것

```python
df #환자들의 나이들을 10살 단위로 나눠보자!
```

```python
#나이가 몇살부터 몇살까지 있는지부터 확인하자
df['age'].describe() #20대부터 60대로 분류
```

```python
pd.cut(df['age'], bins=[20, 30, 40, 50, 60]) #age컬럼을 해당단위로 나눠보자
```

```python
df['age_group'] = pd.cut(df['age'], bins=[20, 30, 40, 50, 60])
#20살에는 구간값이 NAN값이 있는데 확인해보자
#구간값에 (는 포함되지않고 ]는 포함되는 값, 즉 나누는 기준이 애매하다
```

```python
df
```

```python
pd.cut(df['age'], bins=[20, 30, 40, 50, 60], right=False)
#right는 오른쪽값의 포함여부를 말함, 왼쪽값은 포함하고 오른쪽값은 포함하지않는 구간을 설정함
```

```python
df['age_group'] = pd.cut(df['age'], bins=[20, 30, 40, 50, 60], right=False)
```

```python
df #이상-미만으로 값들이 제대로 설정됨,
#하지만 각 구간들이 보기가 어려운데.. 어떻게 하면 좋을까?
```

```python
df['age_group'] = pd.cut(df['age'],
                         bins=[20, 30, 40, 50, 60],
                         right=False,
                         labels=['20대', '30대', '40대', '50대'])
```

```python
df
```

#실습 설명
##환자들의 비만도를 BMI 값에 따라 분류하려고 합니다. 아래 기준에 따라 비만도를 분류해서 obesity라는 컬럼에 저장해 주세요.

##* 18.5 미만(저체중): under
##* 18.5 이상 25 미만(정상): healthy
##* 25 이상 30 미만(과체중): over
##* 30 이상(비만): obese

```python
df
```

```python
BMI_max = df['BMI'].max() + 1
df['obesity'] = pd.cut(df['BMI'],
                               bins=[0, 18.5, 25, 30, BMI_max],
                               labels=['under', 'healthy', 'over', 'obese'],
                               right=False)
df
```
